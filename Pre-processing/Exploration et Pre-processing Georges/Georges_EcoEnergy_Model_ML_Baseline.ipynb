```python
import pandas as pd
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
# import autosklearn
from sklearn.metrics import accuracy_score, f1_score
import os
```

```python
path = "/content/drive/MyDrive/Ecoenergy/Results/"

dred = "new_dred_01_20230516.xlsx"
qud = "new_qud_01_20230516.xlsx"
 

# importer le dataset
rw_dred =  pd.read_excel( path + dred)
rw_qud  =  pd.read_excel( path + qud)


rw_dred.head()
```

```python
qud_feature = rw_qud.iloc[ : , 1:-5 ].drop( 'out' , axis = 1 )
qud_target  = rw_qud.iloc[ : , 1:-5 ].out

qud_features_train = qud_feature.iloc[ : int(qud_feature.shape[0]*0.8 ) , : ]
qud_features_test  = qud_feature.iloc[ int(qud_feature.shape[0]*0.8 ) : , : ]  

qud_target_train = qud_target.iloc[ : int(qud_feature.shape[0]*0.8 ) ]
qud_target_test  = qud_target.iloc[ int(qud_feature.shape[0]*0.8 ) : ]  
```

```python
dred_feature = rw_dred.iloc[ : , 1:-3 ].drop( 'out' , axis = 1 )
dred_target  = rw_dred.iloc[ : , 1:-3 ].out

dred_features_train = dred_feature.iloc[ : int(dred_feature.shape[0]*0.8 ) , : ]
dred_features_test  = dred_feature.iloc[ int(dred_feature.shape[0]*0.8 ) : , : ]  

dred_target_train = dred_target.iloc[ : int(dred_feature.shape[0]*0.8 ) ]
dred_target_test  = dred_target.iloc[ int(dred_feature.shape[0]*0.8 ) : ]  
```

Machine Learning (ML)

```python
# qud

xgb = XGBClassifier()


xgb.fit(qud_features_train, qud_target_train)
yhat = xgb.predict(qud_features_test)

print(f"Les scores de chaque classe: \n{f1_score(qud_target_test, yhat, average = None)}")

print()
macro = f1_score(qud_target_test, yhat, average = "macro")
print(f"Les scores de ENSEMBLE classe: \n{macro}")
```

```python
# dred

xgb = XGBClassifier()


xgb.fit(dred_features_train, dred_target_train)
yhat = xgb.predict(dred_features_test)


print(f"Les scores de chaque classe: \n{f1_score(dred_target_test, yhat, average = None)}")

print()
macro = f1_score(dred_target_test, yhat, average = "macro")
print(f"Les scores de ENSEMBLE classe: \n{macro}")

```

```python
# les donnees provenant de simulated dataset

all_files_list = os.listdir(path)

recherche = "new_subsim" # fichiers de `simulated`
files_sim = [element.replace(".xlsx", "") for element in all_files_list if recherche in element]


all_data = {fichier : pd.read_excel( path + f"{fichier}.xlsx") for fichier in files_sim }


for fichier, data in all_data.items() :
  # print(fichier)
  tmp = data.iloc[: , 1:-5]
  input = tmp.drop(["out"] , axis=1)
  output = data["out"]
  if data["out"].nunique() == 5:
    print(f"{fichier}==================== Classes OK")
    ## separation en train et test
    datas_train = input.iloc[ : int(input.shape[0]*0.8 ) , : ]
    datas_test  = input.iloc[ int(input.shape[0]*0.8 ) : , : ]  

    output_train = output.iloc[ : int(input.shape[0]*0.8 ) ]
    output_test  = output.iloc[ int(input.shape[0]*0.8 ) : ]  

    xgb = XGBClassifier()


    xgb.fit(datas_train, output_train)
    yhat = xgb.predict(datas_test)

    print(f"Les scores de chaque classe: \n{f1_score(output_test, yhat, average = None)}")

    print()
    macro = f1_score(output_test, yhat, average = "macro")
    print(f"Les scores de ENSEMBLE classe: \n{macro}\n\n")
    
  else:
    print(f"{fichier}==================== Absence de certaines classes\n\n")
```
